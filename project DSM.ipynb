```python
# Assuming 'Date' column is already in datetime format
# Extract month from 'Date' column and create a new 'Month' column
df['Month'] = df['Date'].dt.month

# Group by month and sum the sales
monthly_sales = df.groupby('Month')['Weekly_Sales'].sum()

# Group by month and sum the weekly sales
monthly_sales = df.groupby('Month')['Weekly_Sales'].sum()

# Plotting month-wise sales
plt.figure(figsize=(14, 7), dpi=88)
plt.bar(monthly_sales.index, monthly_sales.values)
plt.xlabel('Months')
plt.ylabel('Weekly Sales')
plt.title('Monthwise Sales')
plt.show()
```

```python
 SuperBowl=['2010-02-12', '2011-02-11', '2012-02-10']
LabourDay=['2010-09-10', '2011-09-09', '2012-09-07']
ThanksGiving=['2010-11-26', '2011-11-25']
Christmas=['2010-12-31', '2011-12-30']
```

```python
Super_Bowl_sales = df.loc[df['Date'].isin(SuperBowl), 'Weekly_Sales'].mean()
Super_Bowl_sales_rounded = round(Super_Bowl_sales, 2)

Labour_Day_sales = df.loc[df['Date'].isin(LabourDay), 'Weekly_Sales'].mean()
Labour_Day_sales_rounded = round(Labour_Day_sales, 2)

Thanksgiving_sales = df.loc[df['Date'].isin(ThanksGiving), 'Weekly_Sales'].mean()
Thanksgiving_sales_rounded = round(Thanksgiving_sales, 2)

Christmas_sales = df.loc[df['Date'].isin(Christmas), 'Weekly_Sales'].mean()
Christmas_sales_rounded = round(Christmas_sales, 2)

print("Super Bowl Sales:", Super_Bowl_sales_rounded)
print("Labour Day Sales:", Labour_Day_sales_rounded)
print("Thanksgiving Sales:", Thanksgiving_sales_rounded)
print("Christmas Sales:", Christmas_sales_rounded)
```

```python
# Extract year from 'Date' column and create a new 'Year' column
df['Year'] = df['Date'].dt.year

# Group by year and sum the weekly sales
yearly_sales = df.groupby('Year')['Weekly_Sales'].sum()

# Plotting year-wise sales
plt.figure(figsize=(14, 7), dpi=88)
plt.bar(yearly_sales.index, yearly_sales.values)
plt.xlabel('Year')
plt.ylabel('Weekly Sales')
plt.title('Yearwise Sales')
plt.show()
```

```python
# Calculate mean sales for non-holiday weeks
non_holiday_mean_sales = df[df['Holiday_Flag'] == 0]['Weekly_Sales'].mean()

# Find holidays with higher sales than the mean sales in non-holiday weeks
high_sales_holidays = df[(df['Holiday_Flag'] == 1) & (df['Weekly_Sales'] > non_holiday_mean_sales)]['Date'].unique()
print("Holidays with higher sales than non-holiday mean sales:", high_sales_holidays)
```

```python
df['Year'] = pd.to_datetime(df['Date']).dt.year

# Group bymonth and year andsum the weekly saless 
monthly_sales = df.groupby(['Year', 'Month'])['Weekly_Sales'].sum()

# Plot GROUB monthly sales
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='bar')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Monthly Sales')
plt.show()

# Group the data by every 4 months
df['3_Months'] = (df['Month'] - 1) // 3 + 1
four_monthly_sales = df.groupby(['Year', '3_Months'])['Weekly_Sales'].sum()

# Plot every 4 months sales
plt.figure(figsize=(10, 5))
four_monthly_sales.plot(kind='bar')
plt.xlabel('3 Months')
plt.ylabel('Total Sales')
plt.title('Sales semster Every 4 Months')
plt.show()

```

```python
# Scatter plots
plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.scatter(df['Temperature'], df['Weekly_Sales'], color='blue', alpha=0.5)
plt.xlabel('Temperature')
plt.ylabel('Weekly Sales')
plt.title('Temperature vs. Weekly Sales')

plt.subplot(2, 2, 2)
plt.scatter(df['Fuel_Price'], df['Weekly_Sales'], color='green', alpha=0.5)
plt.xlabel('Fuel Price')
plt.ylabel('Weekly Sales')
plt.title('Fuel Price vs. Weekly Sales')

plt.subplot(2, 2, 3)
plt.scatter(df['CPI'], df['Weekly_Sales'], color='red', alpha=0.5)
plt.xlabel('CPI')
plt.ylabel('Weekly Sales')
plt.title('CPI vs. Weekly Sales')

plt.subplot(2, 2, 4)
plt.scatter(df['Unemployment'], df['Weekly_Sales'], color='orange', alpha=0.5)
plt.xlabel('Unemployment')
plt.ylabel('Weekly Sales')
plt.title('Unemployment vs. Weekly Sales')

plt.tight_layout()
plt.show()

```
